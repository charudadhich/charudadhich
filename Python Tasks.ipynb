Problem-1: Stack

```python
class Stack:
    
    # Constructor
    def __init__(self):
        self.stack = []      #create a empty list
        self.top = -1
        self.size = 0
    
    # Adding element to stack and incremting top by 1
    def push(self, item):
        self.stack.append(item)
        self.top += 1
        self.size += 1
        print("The item pushed is:", item)
        
    # Removing elements from stack and decrementin top by 1
    def pop(self):
        if self.top < 0:
            print("Underflow - stack is empty")
        else:    
            item = self.stack.pop()
            self.top -= 1
            self.size -= 1
            print ("Popped element is:", item)
    
    # Return the top element
    def peek(self):
        if self.isEmpty():
            print("entered")
            print("Stack empty, so peek not possible")
        else:
            print ("Top element is:", self.stack[-1])
    
    # Check if stack is empty or not
    def isEmpty(self):
        if self.size == 0:
            print("Stack is empty")
        else:
            print("Stack is not empty")
    
    
    # Display stack elements
    def show(self):
        print('Stack elements are:')
        print(self.stack)
    
newStack = Stack()
newStack.isEmpty()
newStack.push(1)
newStack.push(2)
newStack.show()
newStack.peek()
newStack.pop()    
newStack.show()
newStack.peek()
newStack.pop()
newStack.show()
newStack.pop()
newStack.isEmpty()
```

Problem-2: Queue

```python
class Queue:

    def __init__(self):
        self.queue = []
        self.top = -1
        self.size = 0

    # Add an element
    def enqueue(self, item):
        self.queue.append(item)
        self.top += 1
        self.size +=1
        print("Element enqueued is: ", item)
        
        
    # Remove an element
    def dequeue(self):
        if len(self.queue) < 1:
            print("Already empty queue")
        else:
            item = self.queue.pop(0)
            self.top -= 1
            self.size -= 1
            print("Popped element is: ", item)
            
            
            
    # Return the top element
    def peek(self):
        if self.size == 0:
            print("Queue empty, so peek not possible")
        else:
            print ("Top element is:", self.queue[0])
    
        
    # Check if queue is empty or not
    def isEmpty(self):
        if self.size == 0:
            print("Queue is empty")
        else:
            print("Queue is not empty")
    

    # Display  the queue
    def display(self):
        print(self.queue)

    def size(self):
        return len(self.queue)


q = Queue()
q.display()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.enqueue(5)
q.display()
q.peek()
q.display()
q.dequeue()
print("After removing an element")
q.display()
```

Problem-1: Anagram

```python
def isAnagram(str1, str2):
    if (len(str1) != len(str2)):
        return False              # length of both strings not equal -> not anagram
    
    map = {}

    for i in range(len(str1)):
        if (str1[i] in map):
            map[str1[i]] += 1     # if char of str1 is in map, then increment val of that key by 1
 
        else:
            map[str1[i]] = 1      # if not in map, then insert that key and set value to 1
 
    for i in range(len(str2)):
        if (str2[i] in map):
            map[str2[i]] -= 1     # if char of str2 is in map, then decrement by 1
        else:
            return False          # if char of string not in map, return false

    keys = map.keys()
    for key in keys:
        if (map[key] != 0):       # if all values in map are not zero, then not anagram
            return False

    return True


str1 = input("Enter first string:")
str2 = input("Enter another string:")

if (isAnagram(str1, str2)):
    print("The two strings are anagram of each other")
else:
    print("The two strings are not anagram of each other")
```
